```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
X,y = make_moons(n_samples=500, noise=0.5, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
plt.plot(X[:,0][y==0], X[:,1][y==0], 'yo',alpha=0.5)
plt.plot(X[:,0][y==1], X[:,1][y==1], 'bs',alpha=0.5)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rndcls = RandomForestClassifier(n_estimators=10, random_state=0)
rndcls.fit(X_train, y_train)
y_pred = rndcls.predict(X_test)
y_pred_train = rndcls.predict(X_train)
print(f':{accuracy_score(y_train, y_pred_train)},:{accuracy_score(y_test, y_pred)}')
```
